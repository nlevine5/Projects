```python
from tkinter import *
import requests
import json
import random

url = "https://the-trivia-api.com/api/questions?categories=film_and_tv&limit=10&region=US&difficulty=easy"
response = requests.get(url).json()

possible_q = []

for i in range (0,10):
    possible_q.append(i)

score = 0

def get_qna(response):
    global answer
    global radio
    global answers
    global possible_q
    
    i = random.choice(possible_q)
    possible_q.remove(i)
    question = response[i]['question']
    answer = response[i]['correctAnswer']
    incorrect = response[i]['incorrectAnswers']
    #print(answer)
    
    answers = []
    answers.append(answer)
    for i in incorrect:
        answers.append(i)
    random.shuffle(answers)  #https://www.tutorialspoint.com/How-to-shuffle-a-list-of-objects-in-Python
    radio = IntVar(value=-1) 
    for i in range(0,4):
        x = answers[i]
        index = answers.index(x)
        i = Radiobutton(root, variable=radio, value=index, command=selection)
        i.pack(anchor=N)
        i.config(text=x, wraplength=300)

    q_label.config(text=question)
    
def selection(): 
    global resp_label
    global num_correct
    global score
    resp_label = Label(root, font="Garamond", wraplength=400)
    resp_label.pack()
    selected = radio.get()
    if selected == answers.index(answer):
        resp_label.config(text="Correct!")
        score += 1
    else:
        resp_label.config(text=f"Incorrect! The Correct answer is:\n {answer}")
        
    score_label.config(text=f"Score: {score}")
        
def new_q():
    if bool(possible_q): #https://flexiple.com/python/check-if-list-is-empty-python/
        for widget in root.winfo_children(): #https://www.tutorialspoint.com/getting-every-child-widget-of-a-tkinter-window
            if widget.winfo_class() == 'Radiobutton': #https://tkdocs.com/tutorial/concepts.html
                widget.destroy()

        resp_label.destroy()

        get_qna(response)
    else:
        for widget in root.winfo_children(): 
            if widget.winfo_class() == 'Radiobutton': 
                widget.destroy()
        new_q_button.destroy()
        resp_label.destroy()
        q_label.destroy()
        score_label.destroy()
        final_score.config(text=f"Game Over\nYou got {score}/10 correct!")
        final_score.pack()

root = Tk()
root.geometry("500x500")

new_q_button = Button(root, font=("Garamond", 14), text="Generate New Question", command=new_q)
new_q_button.pack()

score_label = Label(root, font="Garamond", text = f"Score: {score}")
score_label.pack()

q_label = Label(root, font="Garamond", wraplength=400)
q_label.pack()

final_score = Label(root, font="Garamond", wraplength = 400)

get_qna(response)

root.mainloop()

```

The first task I wanted to conquer was generating a new question. First I needed to make a button that would attach to a code that would allow for the tkinter to be reset. At first I was thinking I should just delete every thing and then place nthe widgets in all over again. But the I stumbled upon documentation for 'winfo_children' which allowed me to make the new_q() function. Next I aimed to generate the score. This was fairly easy because I have made many programs before that involve adding 1 to a count when something happens. In this case it is when the user gets a quesiton correct. Then, I aimed to make it so when a quesiton is asked, it does not appear again. I did this through the 'possible_q' list of integers that match up with the responses from the API request. Every time one of the responses is used, that index is deleted from the list 'possible_q' until it is empty. Finally, I wanted to make it so your final score pops up in the end. To do this, I added an 'if/else' statement to my new_q() function that will check if the 'possible_q' list is empty. If it is empty, it destroys all the widgets and displays your score out of 10.

Throughout this project, the hardest part was just getting the first question to work. Pretty much all of the problems I had afterwards were solved by adding 'global' before the variables. Furthermore, I wasn't sure if you wanted the score to be displayed throughout or not, but I decided to add a label to the topthat says your running score. All of the websites I used for help have been linked next to the prts where I used them in the code above.
